```python
def summary_report():
    """
    returns a cost summary report for the project
    """
    import pandas as pd
    A = pd.read_csv('Приложение А.txt', sep='\t')
    print('***')
    print("Общие трудозатраты ", A['Часы'].sum(), 'часов')
    print('***')
    print('Среднее время, затраченное на решение задач в часах \n', A.groupby(['Задача'])['Часы'].mean())
    print('***')
    print('Медианное время, затраченное на решение задач в часах \n', A.groupby(['Задача'])['Часы'].median())
    print('***')
    print('Среднее время, затраченное на решение задач каждым из исполнителей в часах \n', 
         A.pivot_table(['Часы'], ['Исполнитель'], aggfunc='mean'))
    print('***')
    C = pd.read_csv('Приложение C.txt', sep='\t')
    income = 24000
    costs = C[C['Исполнитель']=='Вася']['Ставка'].sum()*A[A['Исполнитель']=='Вася']['Часы'].sum() + \
            C[C['Исполнитель']=='Маша']['Ставка'].sum()*A[A['Исполнитель']=='Маша']['Часы'].sum() + \
            C[C['Исполнитель']=='Петя']['Ставка'].sum()*A[A['Исполнитель']=='Петя']['Часы'].sum()
    profit = income - costs
    print('Pентабельность проекта', profit, 'руб.')
    print('***')
    print('Cреднее количество часов, отрабатываемое каждым сотрудником за день')
    print('--Вася')
    print(A[A['Исполнитель'] == 'Вася'].groupby(['Дата'])['Часы'].mean())
    print('--Маша')
    print(A[A['Исполнитель'] == 'Маша'].groupby(['Дата'])['Часы'].mean())
    print('--Петя')
    print(A[A['Исполнитель'] == 'Петя'].groupby(['Дата'])['Часы'].mean())
    print('***')
    print('Дни отсутствия для каждого сотрудника, с учётом того, что \n все исполнители обязаны были работать каждый будний день')
    A_date = pd.read_csv('Приложение А.txt', sep='\t', parse_dates=['Дата'], dayfirst=True, index_col='Дата' )
    A_date.loc[:,'weekday'] = A_date.index.weekday
    dat_v = []
    for i in range(len(A_date[A_date['Исполнитель']=='Вася']['weekday'])):
        dat_v.append(A_date[A_date['Исполнитель']=='Вася']['weekday'][i])
    for m in range(len(dat_v)-1):
        if dat_v[m]==dat_v[m+1] or dat_v[m]==dat_v[m+1]-1 or (dat_v[m]==4 and dat_v[m+1]==0):
            continue
        else:
            print(A_date[A_date['Исполнитель']=='Вася'].iloc[m:m+2,:])
            print('Есть пропуск!')
    
    dat_p = []
    for i in range(len(A_date[A_date['Исполнитель']=='Петя']['weekday'])):
        dat_p.append(A_date[A_date['Исполнитель']=='Петя']['weekday'][i])
    for m in range(len(dat_p)-1):
        if dat_p[m]==dat_p[m+1] or dat_p[m]==dat_p[m+1]-1 or (dat_p[m]==4 and dat_p[m+1]==0):
            continue
        else:
            print(A_date[A_date['Исполнитель']=='Петя'].iloc[m:m+2,:])
            print('Есть пропуск!')
    
    dat_m = []
    for i in range(len(A_date[A_date['Исполнитель']=='Маша']['weekday'])):
        dat_m.append(A_date[A_date['Исполнитель']=='Маша']['weekday'][i])
    for m in range(len(dat_m)-1):
        if dat_m[m]==dat_m[m+1] or dat_m[m]==dat_m[m+1]-1 or (dat_m[m]==4 and dat_m[m+1]==0):
            continue
        else:
            print(A_date[A_date['Исполнитель']=='Маша'].iloc[m:m+2,:])
            print('Есть пропуск!')
    print('***')
    Bfl = pd.read_csv('Приложение B.txt', sep='\t', index_col='Задача')
    f1 = A[A['Исполнитель'] == 'Вася'].groupby(['Задача'])['Часы'].sum()
    f2 = A[A['Исполнитель'] == 'Петя'].groupby(['Задача'])['Часы'].sum()
    f3 = A[A['Исполнитель'] == 'Маша'].groupby(['Задача'])['Часы'].sum()
    fly = pd.concat([Bfl, f1], axis=1) #Вася
    fly = fly.dropna()
    fly['fly'] = (fly['Часы'] - fly['Оценка'])*100/fly['Оценка']
    fly1 = pd.concat([Bfl, f2], axis=1) #Петя
    fly1 = fly1.dropna()
    fly1['fly'] = (fly1['Часы'] - fly1['Оценка'])*100/fly1['Оценка']
    fly2 = pd.concat([Bfl, f3], axis=1) #Маша
    fly2 = fly2.dropna()
    fly2['fly'] = (fly2['Часы'] - fly2['Оценка'])*100/fly2['Оценка']
    print(round(fly['fly'].mean(), 2), '% средний вылет специалиста Васи из оценки')
    print(round(fly1['fly'].mean(), 2), '% средний вылет специалиста Пети из оценки')
    print(round(fly2['fly'].mean(), 2), '% средний вылет специалиста Маши из оценки')
    print('***')
    from matplotlib import pyplot as plt
    params = { 
    'axes.labelsize': "large",
    'xtick.labelsize': 'x-large',
    'legend.fontsize': 20,
    'figure.dpi': 150,
    'figure.figsize': [10, 4]
    }
    plt.rcParams.update(params)
    Bfl = pd.read_csv('Приложение B.txt', sep='\t', index_col='Задача')
    new = pd.concat([Bfl,A.groupby(['Задача'])['Часы'].mean()], axis = 1)
    ax = new.plot.bar(rot=0, color = ['g','r'])
    #plt.xticks(list(range(1,38,2)), [list(new.index[::2])])
    plt.xlabel('Номер задачи')
    plt.ylabel('Часы')
    plt.savefig('saved_figure.png')
    
    
```

```python
summary_report()
```
